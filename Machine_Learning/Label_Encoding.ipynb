Label Encoding:

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.datasets import load_breast_cancer
```

```python
#Label Encoding of Breast Cancer Dataset
df = load_breast_cancer()
```

```python
df = pd.read_csv('/content/data.csv')
```

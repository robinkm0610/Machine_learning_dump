```python
from matplotlib import pyplot as plt
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data[0]
```

```python
plt.gray()
for i in range(5):
  plt.matshow(digits.images[i])
```

```python
digits.target[0:5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data,digits.target, test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression 
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict(X_test)
```

```python
plt.matshow(digits.images[67])
```

```python
digits.target[67]
```

```python
model.predict([digits.data[67]])
```

```python
y_predit = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predit)
cm
```

```python
import seaborn as sns
plt.figure(figsize = (10,7))
sns.heatmap(cm, annot= True)
plt.xlabel('predicted')
plt.ylabel('Truth')
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
import pandas as pd


df = pd.DataFrame(iris.data,columns = iris.feature_names )
df['target'] = pd.Series(iris.target)
df.head(5)
```

```python
df.rename(columns = 
{'sepal length (cm)':'sepal_length','sepal width (cm)':'sepal_width' ,'petal length (cm)':'petal_length','petal width (cm)':'petal width'}, inplace = True
)
df.head(5)
```

```python
df.rename(columns = {'petal width':'petal_width'}, inplace = True)
```

```python
df.head(5)
```

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.model_selection import train_test_split
```

```python
X = df[['sepal_length','sepal_width','petal_length','petal_width']]
y = df.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict(X_test)
```

```python
import pandas as pd
```

```python
df = pd.read_csv(r"homeprices1.csv")
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df,dummies], axis = 'columns')
merged
```

```python
# multivariable convariance - so we have to drop one of the dummy variables to avoid multivariable trap
final = merged.drop(['town','west windsor'], axis = 1)
```

```python
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = final.drop('price', axis = 1)
```

```python
X
```

```python
y = final.price
y
```

```python
model.fit(X,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(X,y) #checking the accuracy
```

```python
import pandas as pd
df = pd.read_csv("carprice.csv")
df
```

```python
dummies = pd.get_dummies(df['Car Model'])
dummies 
```

```python
merged = pd.concat([df,dummies], axis = 1)
merged
```

```python
final = merged.drop(['Car Model','Mercedez Benz C class'], axis = 1)
final
```

```python
X = final.drop('Sell Price($)', axis = 1)
X
```

```python
y = final['Sell Price($)']
y
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(X,y)
```

```python
reg.score(X,y)
```

```python
#Price of mercedez benz that is 4 yr old with mileage 45000

reg.predict([[45000,4,0,0]])
```

```python
#Price of BMW X5 that is 7 yr old with mileage 86000
reg.predict([[86000,7,0,1]])
```

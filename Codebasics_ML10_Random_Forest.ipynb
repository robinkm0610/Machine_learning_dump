```python
import pandas as pd
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
dir(digits)
```

```python
df =pd.DataFrame(digits.data)
df.head()
```

```python
df['target'] = digits.target
```

```python
df
```

```python
from matplotlib import pyplot as plt
plt.gray()
for i in range(4):
    plt.matshow(digits.images[i])
    
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(['target'],axis = 1),digits.target, test_size = 0.2)
```

```python

from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators= 60)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot = True)
plt.xlabel('predict')
plt.ylabel('Truth')
```

```python
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
```

```python
dir(iris)
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df
```

```python
df['target'] = iris.target
df
```

```python
df.rename({'sepal length (cm)':'sepal_length',	'sepal width (cm)':'sepal_width','petal length (cm)':'petal_length','petal width (cm)':'petal_width'},inplace = True, axis = 1)
df
```

```python
iris.target_names
```

```python
df['flower_name'] = df.target.apply(lambda x : iris.target_names[x])
df
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators= 40)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(['target'	,'flower_name'], axis = 1),df.target, test_size = 0.3)
```

```python
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

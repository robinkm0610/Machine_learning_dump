```python
import pandas as pd
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
target = df.Survived
inputs = df.drop('Survived', axis = 1)
inputs.head()
```

```python
dummies = pd.get_dummies(inputs.Sex)
dummies.head()
```

```python
inputs = pd.concat([inputs, dummies],axis = 1)
inputs.head()
```

```python
inputs.drop('Sex', axis = 1, inplace = True)
inputs.head()
```

```python
#I am dropping male column as well because of dummy variable trap theory. One column is enough to repressent male vs female
inputs.drop('male', axis = 1, inplace = True)
inputs.head()
```

```python
inputs.isna().sum()
```

```python
inputs.Age.fillna(df.Age.mean(),inplace = True)
```

```python
inputs.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size = 0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
#model.fit(X_train , y_train)

model.fit(X_train,y_train)
model.score(X_test, y_test)
```

```python
#calculate the score using cross_validation
from sklearn.model_selection import cross_val_score
cross_val_score(GaussianNB(), X_train, y_train, cv = 5)
```

```python
df = pd.read_csv('email.csv')
df.head(5)
```

```python
df.Category.value_counts()
```

```python
df['spam'] = df.Category.apply(lambda x : 1 if x == 'ham' else 0)
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.Message, df.spam, test_size = 0.25)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v= CountVectorizer()
x_train_count = v.fit_transform(X_train.values)
x_train_count.toarray()[:2]
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train_count, y_train)
```

```python
emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]
emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
X_test_count = v.transform(X_test)
model.score(X_test_count, y_test)
```

### Sklearn pipeline

```python
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python

clf.fit(X_train, y_train)
```

```python

clf.score(X_test,y_test)
```

```python
clf.predict(emails)
```

```python
from sklearn import datasets
wine = datasets.load_wine()
```

```python
df['target'] = wine.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('target', axis = 1),df.target, test_size = 0.25)

```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
mn = MultinomialNB()
mn.fit(X_train,y_train)
mn.score(X_test,y_test)
```
